```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
import re
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
path =  "C:/Users/farim/PycharmProjects/ChatGPT_Tutor_Project/"

#with open(path+"Documents/emb_02450", "rb") as fp:
#    embeddings = pickle.load(fp)

df = pd.read_pickle('./Data_Generation/df_pickle/final_02450_emb.pkl')
# load the embeddings
    # make the df only contain the unique contexts
df = df.drop_duplicates(subset=['context'])
emb_np = np.stack(df['context_embedding'].to_numpy())

```

```python
emb_np[0]
```

```python
tsne_model = TSNE(perplexity=40, n_components=3, init='pca', n_iter=2500, random_state=23)
new_values = tsne_model.fit_transform(emb_np)

x = []
y = []
z=[]
for value in new_values:
    x.append(value[0])
    y.append(value[1])
    z.append(value[2])

colors=np.arange(len(x))
```

```python
#2d plot
plt.scatter(x,y,c=colors,cmap='jet')
plt.colorbar()


# display the plot
plt.show()
```

```python
#3d plot

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=colors, cmap='jet')

# display the plot
plt.show()
```
